# Netflix Stock Sales Analysis

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
sns.set(rc={'figure.figsize' : (10,5)})
```

```python
df = pd.read_csv("Netflix.csv")
```

```python
df.head()
```

# Setting Date column as index column

```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')
df.head()
```

# Volume of Strock vs Time

```python
sns.lineplot(x=df.index, y=df.Volume, label = 'volume')
plt.title('Volume of stock vs Time')
```

# Stocks Prices High Open Close

```python
df.plot(y=['High','Close','Open'], title = 'Netflix Stock Price')
```

# Stock Prices Volume Daily Monthly Yearly

```python
fig, (ax1,ax2,ax3) = plt.subplots(3, figsize=(20,25))
df.groupby(df.index.day).mean().plot(y='Volume', ax = ax1, xlabel = 'Day')
df.groupby(df.index.month).mean().plot(y='Volume', ax = ax2, xlabel = 'Month')
df.groupby(df.index.year).mean().plot(y='Volume', ax = ax3, xlabel = 'Yearly')
```

# Top 5 Dates with High opening 

```python
a = df.sort_values(by='High', ascending = False).head()
a['High']
```

# Top 5 Dates with Low opening 

```python
a = df.sort_values(by='Low', ascending = True).head()
a['Low']
```

# High & Low Values Stock per period of time

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2, sharex = True, figsize = (12,5))
fig.suptitle('High & Low Values Stock per period of time', fontsize = 18)
sns.lineplot(ax=axes[0], y = df['High'],x=df.index, color='green')
sns.lineplot(ax=axes[1], y = df['Low'],x=df.index, color='red')
```
